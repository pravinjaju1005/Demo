```python
df= pd.read_csv('MegaMart_sales.csv')
a= pd.DataFrame(df)
df_new= pd.read_csv('MegaMart_newsales.csv')
b = pd.DataFrame(df_new)
#a.groupby(by='Category').Sales.sum()
# b.groupby(by='Category').Sales.sum()
```

```python
z = pd.concat([a,b])
#z.drop_duplicates(subset=('Sales'),keep=False,inplace=True)
z.groupby(by='Sub-Category').Profit.sum()
#z.pivot_table(index='Category', columns=None,values='Sales',aggfunc='sum')
```

```python
z.drop_duplicates(subset=None,keep='first',inplace = True)
z.groupby(by='Category').Sales.sum()
```

```python
def PQ(x):
    a = x['Profit']
    b = x['Quantity']
    c = a/b
    return c

z['PQ']= z.apply(PQ,axis=1)
z.sort_values('PQ',ascending=False).head(52)
```
