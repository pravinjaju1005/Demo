## dropna()

Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null values in different ways.<br>
**Syntax - DataFrameName.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)**<br>
**Crucial Parameters**-<br>
axis: axis takes int or string value for rows/columns. Input can be 0 or 1 for Integer and ‘index’ or ‘columns’ for String.<br>
**how: how takes string value of two kinds only (‘any’ or ‘all’). ‘any’ drops the row/column if ANY value is Null and ‘all’ drops only if ALL values are null.**<br>
inplace: It is a boolean which makes the changes in data frame itself if True.<br>

```python
#dropping the records
nba.dropna(how='all',inplace=True)# will remove null value only if all the colum contains null as how vaue is 'all'with axis 0.
```

if any col contains null : any
if all the col contain null : all

##  .sample() - for sampling purpose
Return a random sample of items from an axis of object.<br>
syntax is - bond.sample(
    n=None,
    frac=None,
    replace=False,
    weights=None,
    random_state=None,
    axis=None,
) -> ~FrameOrSeries
:
<br>

n - The n parameter enables you to specify how many items to return in the sample.<br>
FRAC-The frac parameter enables you to specify the fraction of items to return in the sample.The argument to this parameter should be a floating point number.So for example, if you set frac = .5, then the sample() method will return 50% of the rows from the dataframe.Note that frac can not be used with the n parameter. You need to use one or the other.<br>

REPLACE- By default, replace is set to replace = False. In this case, a row can be sampled only one time, and will not be available for selection again.<br>
AXIS- The axis parameter enables you to specify the axis from which to sample.

## get_group() 
**After calling groupby(), you can access each group dataframe individually using get_group()**.Please note it works only if you have not apllied any other aggregation function with groupby like if write sectors = fortune.groupby('Sector').sum() - then get_group() won;t work.

**Syntax --> dataframename.get_group("groupname like if we grouped on the basis of sector then one sector name here")**

```python
sectors = fortune.groupby('Sector')
```

```python
sectors.get_group('Technology')
```

```python
sectors.get_group('Apparel')['Profits'].sum()
```

```python
sectors.mean()
```
